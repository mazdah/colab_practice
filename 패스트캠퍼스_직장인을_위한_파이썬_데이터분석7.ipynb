```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 1000
```

```python
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

---
- seaborn에서는 size와 sizes를 동시에 지정한다.
- sizes 옵션에는 사이즈의 min과 max를 명시해준다.
- hue는 컬러 옵션이다.
- palette를 이용하여 색상을 지정한다.
---

```python
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), hue=area, palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(131)
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), color='blue', alpha=0.1)
plt.title('alpha=0.1')

plt.subplot(132)
plt.title('alpha=0.5')
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), color='red', alpha=0.5)

plt.subplot(133)
plt.title('alpha=1.0')
sns.scatterplot(x, y, size=area, sizes=(area.min(), area.max()), color='green', alpha=0.9)

plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

sns.barplot(x, y, alpha=0.8, palette='YlGnBu')

plt.ylabel('Scores')
plt.title('Subjects')
plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

# x와 y의 순서를 바꾸는 것만으로 barh 역할을 한다.
sns.barplot(y, x, alpha=0.8, palette='YlGnBu')

plt.ylabel('Scores')
plt.title('Subjects')
plt.show()
```

```python
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width=0.35
fig, axes = plt.subplots()

axes.bar(x - width / 2, y_1, width, align='center', alpha=0.5, color='green')
axes.bar(x + width / 2, y_2, width, align='center', alpha=0.8, color='red')

plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Score')
plt.title('Subjects')

plt.legend(['john', 'petter'])
plt.show()
```

# seaborn - violinplot

- 컬럼에 대한 데이터의 비교 분포도를 확인할 수 있음
- 곡선진(뚱뚱한) 부분은 데이터의 분포를 나타냄
- 양쪽 끝의 뾰족한 부분은 데이터의 최솟값과 최댓값을 나타냄
---

```python
sns.violinplot(x=tips['total_bill'])
plt.show()
```

---
- x, y축을 지정해줌으로써 바이올린을 분할하여 비교 분포를 볼 수 있음
---

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.violinplot(y='day', x='total_bill', data=tips)
plt.show()
```

---
- hue 옵션을 주면 단일 컬럼에 대한 바이올린 모양의 비교를 할 수 있다.
- split 옵션과 함께 사용해야 보다 명확한 비교가 가능
---

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, palette='muted', split=True)
plt.show()
```

# seaborn - lmplot

- 컬럼간의 선형 관계를 확인하기에 용이한 챠트
- outlier도 같이 짐작해볼 수 있음
---

```python
sns.lmplot(x='total_bill', y='tip', height=8, data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', height=8, data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='day', col_wrap=2, height=8, data=tips)
plt.show()
```

---
#seaborn- relplot

- 두 컬럼 간의 상관관계를 보여준다.
- lmplot처럼 선형 관계를 보여주지는 않는다.
---

```python
sns.relplot(x='total_bill', y='tip', hue='day', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='day', col='time', data=tips)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='day', row='sex', col='time', data=tips)
plt.show()
```

# seaborn - jointplot

- scatter(산점도)와 histogram(분포)을 동시에 그려준다.
- 수자형 데이터만 표현 가능하다.
---

```python
sns.jointplot(x='total_bill', y='tip', height=8, data=tips)
plt.show()
```

```python
# 선형 관계를 표현하는 regression 라인 그리기 : kind='reg' 옵션
sns.jointplot(x='total_bill', y='tip', height=8, data=tips, kind='reg')
plt.show()
```

```python
# hex 밀도 그리기 : kind='hex' 옵션
sns.jointplot(x='total_bill', y='tip', height=8, data=tips, kind='hex')
plt.show()
```

```python
# 등고선 그리기 : kind='kde' 옵션
iris = sns.load_dataset('iris')
sns.jointplot(x='sepal_width', y='petal_length', height=8, data=iris, kind='kde', color='g')
plt.show()
```
