### 다중 그래프 - fig, axes, subplot

- subplot(row, column, index)
- subplots(행의 갯수, 열의 갯수)
---

```python
data = np.arange(100, 201)
plt.subplot(1, 3, 1)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(1, 3, 2)
plt.plot(data2)

data3 = np.arange(300, 401)
plt.subplot(1, 3, 3)
plt.plot(data3)

plt.show()
```

```python
data = np.arange(100, 201)
plt.subplot(211)
plt.plot(data)

data2 = np.arange(200, 301)
plt.subplot(212)
plt.plot(data2)

plt.show()
```

```python
data = np.arange(1, 51)

fig, axes = plt.subplots(2, 3)

axes[0, 0].plot(data)
axes[0, 1].plot(data * data)
axes[0, 2].plot(data ** 3)
axes[1, 0].plot(data % 10)
axes[1, 1].plot(-data)
axes[1, 2].plot(data // 20)

plt.tight_layout()
plt.show()
```

### barplot, barh
---

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.bar(x, y, align='center', alpha=0.7, color='red')
plt.xticks(x, fontsize=15)
plt.ylabel('Number of Student', fontsize=20)
plt.title('Subjects', fontsize=20)

plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.barh(x, y, align='center', alpha=0.7, color='red')
plt.yticks(x, fontsize=15)
plt.xlabel('Number of Student', fontsize=20)
plt.title('Subjects', fontsize=20)

plt.show()
```

```python
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width = 0.35
fig, axes = plt.subplots()

axes.bar(x - width / 2, y_1, width, align='center', alpha=0.5)
axes.bar(x + width / 2, y_2, width, align='center', alpha=0.8)

plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.legend(['john', 'peter'])
plt.show()
```

```python
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width = 0.35
fig, axes = plt.subplots()

axes.barh(x - width / 2, y_1, width, align='center', alpha=0.5)
axes.barh(x + width / 2, y_2, width, align='center', alpha=0.8)

plt.yticks(x)
axes.set_yticklabels(x_label)
plt.xlabel('Number of Students')
plt.title('Subjects')

plt.legend(['john', 'peter'])
plt.show()
```
