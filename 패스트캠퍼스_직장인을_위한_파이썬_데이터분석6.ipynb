#### 다중 그래프
---

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 3,
                         sharey=True,           # 각 그래프간에 y축을 공유
                         tight_layout=True)     # 그래프의 패딩을 자동으로 조절하여 tight한 그래프를 생성
fig.set_size_inches(12, 5)

axes[0].hist(x, bins=bins)
axes[1].hist(x, bins=bins * 2)
axes[2].hist(x, bins=bins * 4)

plt.show()
```

---
#### Y축에 Density 표기
---

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 2, tight_layout=True)
fig.set_size_inches(9, 3)

axes[0].hist(x, bins=bins, density=True, cumulative=True) # cumulative=True : 누적 분포를 보기 위한 파라미터
axes[1].hist(x, bins=bins, density=True)

plt.show()
```

### Pie Chart

#### Pie chart 옵션
- explode: pie에서 튀어나온 비율
- autopct : 퍼센트 자동으로 표기
- shadow : 그림자 표시
- startangle : 파이를 그리기 시작할 각도

texts, autotexts 인자를 리턴받는다.

**texts** : label에 대한 텍스트 효과

**autotexts** : 파이 위에 그려지는 텍스트 효과

---

```python
labels = ['Samsung', 'Huawai', 'Apple', 'Xiaomi', 'Oppo', 'etc']
size = [20.4, 15.8, 10.5, 9., 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0)

patches, texts, autotexts = plt.pie(size,
                                    explode=explode,
                                    labels=labels,
                                    autopct='%1.1f%%',
                                    shadow=True,
                                    startangle=90)

plt.title('Smartphone pie', fontsize=15)

for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

for t in autotexts:
    t.set_color('white')
    t.set_fontsize(18)

plt.show()
```

---
#### 다중 boxplot 생성
---

```python
# 샘플 데이터 생성
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))

spread = np.random.rand(50) * 100
center = np.ones(25) * 40

flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100

d2 = np.concatenate((spread, center, flier_high, flier_low))

data.shape = (-1, 1)
d2.shape = (-1, 1)

data = [data, d2, d2[::2,0]]
```

```python
plt.boxplot(data)
plt.tight_layout()
plt.show()
```

```python
# 축 바꾸기
plt.title('Horizontal Box Plot', fontsize=15)
# 코드를 입력해 주세요
plt.boxplot(data, vert=False)
plt.tight_layout()
plt.show()
```

```python
ax = plt.axes(projection='3d')

z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot(x, y, z, 'gray')
plt.show()
```

```python
ax = plt.axes(projection='3d')

sample_size = 100
x = np.cumsum(np.random.normal(0, 1, sample_size))
y = np.cumsum(np.random.normal(0, 1, sample_size))
z = np.cumsum(np.random.normal(0, 1, sample_size))

ax.plot3D(x, y, z, alpha=0.6, marker='o')

plt.title('ax.plot')
plt.show()
```

---
### 3D scatter 그리기
---

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111, projection='3d') 

sample_size = 500

x = np.cumsum(np.random.normal(0, 5, sample_size))
y = np.cumsum(np.random.normal(0, 5, sample_size))
z = np.cumsum(np.random.normal(0, 5, sample_size))

ax.scatter(x, y, z, c=z, s=20, alpha=0.5, cmap='Greens')

plt.title('ax.scatter')
plt.show()
```

---
### contour3D 그리기(등고선)
---

```python
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(12, 6))
ax = plt.axes(projection='3d')

ax.contour3D(x, y, z, 20, cmap='Reds')

plt.title('ax.contour3D')
plt.show()
```

### imshow
---

```python
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.images[:10]
X[0]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, sharex=True, figsize=(12, 6), sharey=True)

for i in range(10):
    axes[i//5][i%5].imshow(X[i], cmap='Blues')
    axes[i//5][i%5].set_title(str(i), fontsize=20)

plt.tight_layout()
plt.show()
```
