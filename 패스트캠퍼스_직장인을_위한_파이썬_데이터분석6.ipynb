### histogram
---

```python
N = 100000
bins = 30

x = np.random.randn(N)

plt.hist(x, bins=bins)

plt.show()
```

---
#### 다중 그래프
---

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 3,
                         sharey=True,           # 각 그래프간에 y축을 공유
                         tight_layout=True)     # 그래프의 패딩을 자동으로 조절하여 tight한 그래프를 생성
fig.set_size_inches(12, 5)

axes[0].hist(x, bins=bins)
axes[1].hist(x, bins=bins * 2)
axes[2].hist(x, bins=bins * 4)

plt.show()
```

---
#### Y축에 Density 표기
---

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplots(1, 2, tight_layout=True)
fig.set_size_inches(9, 3)

axes[0].hist(x, bins=bins, density=True, cumulative=True) # cumulative=True : 누적 분포를 보기 위한 파라미터
axes[1].hist(x, bins=bins, density=True)

plt.show()
```

### Pie Chart

#### Pie chart 옵션
- explode: pie에서 튀어나온 비율
- autopct : 퍼센트 자동으로 표기
- shadow : 그림자 표시
- startangle : 파이를 그리기 시작할 각도

texts, autotexts 인자를 리턴받는다.

**texts** : label에 대한 텍스트 효과

**autotexts** : 파이 위에 그려지는 텍스트 효과

---

```python
labels = ['Samsung', 'Huawai', 'Apple', 'Xiaomi', 'Oppo', 'etc']
size = [20.4, 15.8, 10.5, 9., 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0)

patches, texts, autotexts = plt.pie(size,
                                    explode=explode,
                                    labels=labels,
                                    autopct='%1.1f%%',
                                    shadow=True,
                                    startangle=90)

plt.title('Smartphone pie', fontsize=15)

for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

for t in autotexts:
    t.set_color('white')
    t.set_fontsize(18)

plt.show()
```
