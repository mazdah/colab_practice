### ▶︎ 전처리 : lambda, map

- lambda는 1줄로 작성하는 간단한 함수
- return을 별도로 명기하지 않는다.
---

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://bit.ly/ds-korean-idol")
df.head()
```

```python
f = lambda x : 1 if x == "남자" else 0
f
```

```python
df["성별"].apply(f)
```

```python
df["키/2"] = df["키"].apply(lambda x : x / 2)
df
```

```python
my_map = {
    '남자': 1,
    '여자': 0
}
```

```python
df["성별"].map(my_map)
```

```python
my_map = {
    '남자': 'male',
    '여자': 'female'
}
df["성별"].map(my_map)
```

### ▶︎ 전처리 : 원핫인코딩의 개념과 get_dummies

- One-Hot-Encoding : 하나의 요소만 True(1)로 만들고 나머지 요소들은 False(0)으로 만드는 기법
- 수치형 데이터가 있는 경우 AI는 연산 과정에서 각 값들 사이의 관계를 스스로 연관 지을 수 있다.
- 예컨대 아래 blood_map에서 'O' = 'B' + 'AB'로 판단할 수 있다는 의미다.
- 따라서 각각의 값이 아무런 관계도 없이 독립적이라는 것을 표현해줘야 한다.
---

```python
blood_map = {
    'A' : 0,
    'B' : 1,
    'AB' : 2,
    'O' : 3
}

df["혈액형_code"] = df["혈액형"].map(blood_map)
```

```python
df
```

```python
df["혈액형_code"].value_counts()
```

```python
df["혈액형_code"]
```

```python
pd.get_dummies(df["혈액형_code"])
```

```python
pd.get_dummies(df["혈액형_code"], prefix="혈액형")
```

```python
df2 = pd.concat([df, pd.get_dummies(df["혈액형_code"], prefix="혈액형")])
df2
```
